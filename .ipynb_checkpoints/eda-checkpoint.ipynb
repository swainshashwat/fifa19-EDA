#### Cleaning `Value`

```python
# Cleaning Value
def value_parser(x):
    val = x[1:-1]
    mk = x[-1]
    try:
        if mk=='M':
            return float(val)*100.0
        else:
            return float(val)
    except ValueError:
        return np.nan
```

```python
df_new = df['Value'].apply(value_parser)
```

```python
# mean of 'Value' feature
np.ceil(np.mean(df_new))
```

```python
df_new.fillna(np.ceil(np.mean(df_new)), inplace=True)
```

```python
df_new = pd.DataFrame(df_new, columns=['Value'])
```

#### Cleaning `Joined`

```python
print('No. of Nan values in `Joined`:',df['Joined'][df['Joined'].isna()].shape[0])
```

```python
# Converting `joined` to type Datetime object
df_new['Joined'] = pd.to_datetime(df['Joined'])

# getting current date as a pd.Series
this_year = pd.to_datetime(pd.Series(['Jan 1, 2019']*df_new['Joined'].shape[0]))

# Calculating player Exp
df_new['Exp'] = this_year - df_new['Joined']

# Dropping Joined
df_new.drop('Joined', axis=1, inplace=True)

# Converting to String
df_new['Exp'] = df_new['Exp'].astype(str) 
```

```python
def exp_parser(x):
    return x.split(' ')[0]
```

```python
df_new['Exp'] = df_new['Exp'].apply(exp_parser)
```

```python
def exp_to_num(x):
    try:
        return int(x)
    except ValueError:
        return np.nan
```

```python
# cleaning nan values
df_new['Exp'] = df_new["Exp"].apply(exp_to_num)
# filling nan with mean
df_new['Exp'].fillna(np.ceil(np.mean(df_new['Exp'])), inplace=True)
```

#### Cleaning `Weight`

```python
df['Weight'].unique()
```

```python
def wght(x):
    try:
        return int(x[:-3])
    except TypeError:
        return np.nan
```

```python
# parsing `Weight` into interger type
df_new['Weight'] = df['Weight'].apply(wght)
```

```python
# No. of nan values in `Weight` feature
print("No. of NaN values in `Weight` feature is:",df_new['Weight'][df_new['Weight'].isna()].shape[0])
```

```python
# filling nan values
df_new['Weight'].fillna(np.mean(df_new['Weight']), inplace=True)
```

#### Cleaning `Height`

- 1 foot = 12 inches
- 1 inch = 2.54 cm

```python
df['Height'].unique()
```

```python
# parsing Height to cm
def parse_height_to_cm(x):
    try:
        fts, inchs = x.split("'")
        return (int(fts)*12 + int(inchs))*2.54
    except:
        return np.nan
```

```python
df_new['Height'] = df['Height'].apply(parse_height_to_cm)
```

```python
print('Number of Nan values in `Height` is:', df_new['Height'][df_new['Height'].isna()].shape[0])
```

```python
# filling missing values
df_new['Height'] = df_new['Height'].fillna(np.ceil(np.mean(df_new['Height'])))
```
